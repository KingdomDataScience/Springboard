**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_increase_rate.

You'd want this function to:
1. Take a filter of df_clean, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: df_clean[df_clean['London_Borough']=='Hillingdon'].
2. Get the Average Price for that Borough, for the years 1995 and 2020.Here we will assume that the number of house in each London Borough stay the same over the year.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2020.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1995 and 2020.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Let us use the function 'grouby' to help us calculate the mean for each year for each Borough. 
#We will consider that the number of house over the year stayed the same for each borough.
df=df_clean.groupby(['London_Borough', 'Year']).mean()
df.sample(10)
```

```python
# Let's reset the index for our new DataFrame df, and call the head() method on it. 
df = df.reset_index()
df.head()
```

```python
#Create a function that compute the increase rate in average price from 1995 to 2020.
def create_price_increase_rate(data):
    year_1995=float(data['Average Prices'][data['Year']==1995])
    year_2020=float(data['Average Prices'][data['Year']==2020])
    increase_rate=[(year_2020-year_1995)/year_1995]
    return increase_rate
```

```python
df.columns
```

```python
#Now let us therefore compute all the increase rate for all the Boroughs in London and put it in a dictionary called final
final={}
for b in df['London_Borough'].unique():
    borough=df[df['London_Borough']==b]
    final[b]=create_price_increase_rate(borough)

print(final)
```

```python
#Let put our dictionary in a pandas dataframe, transpose it and reset the index
final_pd=pd.DataFrame(final)
final_pd=final_pd.T
final_pd=final_pd.reset_index()
final_pd.head()
```

```python
#Rename the column
final_pd.rename(columns={'index':'Borough', 0: 'Increase Rate'}, inplace=True)
final_pd.head()
```

```python
#Let us sort in descending order and select the top 5 boroughs
top5=final_pd.sort_values('Increase Rate', ascending=False).head(5)
print(top5)
```

```python
#Let us therefore do a bar char that display the boroughs that have seen the greatest changes in prices
ax=top5[['Borough', 'Increase Rate']].plot(kind='bar')
ax.set_xticklabels(top5.Borough)
```
