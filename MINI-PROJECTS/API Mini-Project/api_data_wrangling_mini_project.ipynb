```python
def median(vol):
    vol.sort()
    if len(vol) % 2 == 0:
        median1 = vol[len(vol) // 2]
        median2 = vol[(len(vol) // 2) - 1]
        median = (median1 + median2) / 2
    else:
        median = vol[len(vol) // 2]
    return median
```

```python
print('The median trading volume during the year of 2017 was', median(volume))
```
